```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
import pandas as pd
# 线形回归问题，优化函数为minibatch SGD
rate = 0.2 # learning rate
def da(y,y_p,x):
    return (y-y_p)*(-x)

def db(y,y_p):
    return (y-y_p)*(-1)

def calc_loss(a,b,x,y):
    tmp = y - (a * x + b)
    tmp = tmp ** 2  # 对矩阵内的每一个元素平方
    SSE = sum(tmp) / (2 * len(x))
    return SSE

```

```python
def shuffle_data(x,y):
    # 随机打乱x，y的数据，并且保持x和y一一对应
    seed = random.random()
    random.seed(seed)
    random.shuffle(x)
    random.seed(seed)
    random.shuffle(y)

def get_batch_data(x,y,batch=3):
    shuffle_data(x,y)
    x_new = x[0:batch]
    y_new = y[0:batch]
    return [x_new,y_new]
```

```python
#  pandas模拟数据
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
# x = [30	,35,37,	59,	70,	76,	88,	100]
# y = [1100,	1423,	1377,	1800,	2304,	2588,	3495,	4839]

# 数据归一化
x_max = max(x)
x_min = min(x)
y_max = max(y)
y_min = min(y)

for i in range(0,len(x)):
    x[i] = (x[i] - x_min)/(x_max - x_min)
    y[i] = (y[i] - y_min)/(y_max - y_min)


# 初始化a,b值
a = 10.0
b = -20.0
fig = plt.figure(1, figsize=(20, 8))
```

```python
all_loss = []
all_step = []
last_a = a
last_b = b
for step in range(1,200):     #200 epoch
    loss = 0
    all_da = 0
    all_db = 0
    shuffle_data(x,y)
    [x_new,y_new] = get_batch_data(x,y,batch=4)
    for i in range(0,len(x_new)):
      y_p = a*x_new[i] + b
      loss = loss + (y_new[i] - y_p)*(y_new[i] - y_p)/2    #均方误差损失函数
      #loss = loss+ np.mean(-np.sum(y_p*np.log(y_new)))    #交叉熵损失函数,训练还存在问题,loss过大
      all_da = all_da + da(y_new[i],y_p,x_new[i])
      all_db = all_db + db(y_new[i],y_p)
    #loss_ = calc_loss(a = a,b=b,x=np.array(x),y=np.array(y))
    loss = loss/len(x_new)


    # 绘制一部分loss点
    plt.subplot(2,2,1)
    plt.scatter(a,b,s=5,color='blue')
    plt.plot([last_a,a],[last_b,b],color='red')
    # 绘制回归直线
    plt.subplot(2, 2, 3)
    plt.plot(x, y)
    plt.plot(x, y, 'o')
    x_ = np.linspace(0, 1, 2)
    y_draw = a * x_ + b
    plt.plot(x_, y_draw)
    # 绘制loss更新曲线
    all_loss.append(loss)
    all_step.append(step)
    plt.subplot(2,2,4)
    plt.plot(all_step,all_loss,color='orange')
    plt.xlabel("step")
    plt.ylabel("loss")


    # print('a = %.3f,b = %.3f' % (a,b))
    last_a = a
    last_b = b
    a = a - rate*all_da
    b = b - rate*all_db

    if step%1 == 0:
        print("step: ", step, " loss: ", loss)
        plt.show()
plt.show()
```
